# Pandas DataFrame

```python
import numpy as np
import pandas as pd
```

```python
# 1. Write a Pandas program to get the powers of an array values element-wise.

x = {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]}
pd.DataFrame(x)
```

```python
# 2. Write a Pandas program to create and display a DataFrame from a specified 
# dictionary data which has the index labels.

x = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(x, index=labels)
df
```

```python
# 3. Write a Pandas program to get the first 3 rows of a given DataFrame.

df.head(3)
```

```python
# 4. Write a Pandas program to select the 'name' and 'score' columns from the following DataFrame.

df[['name', 'score']]
```

```python
# 5. Write a Pandas program to select the specified columns and rows from a given data frame.

df['name'].loc[['e', 'f', 'g']]
```

```python
# 6. Write a Pandas program to select the rows where the number of attempts in 
# the examination is greater than 2. 

df[df['attempts'] > 2]
```

```python
# 7. Write a Pandas program to count the number of rows and columns of a DataFrame

df.shape
```

```python
# 8. Write a Pandas program to select the rows where the score is missing, i.e. is NaN.

ind = np.where(df['score'].isnull() == True)
df.iloc[ind]
```

```python
# 9. Write a Pandas program to select the rows the score is between 15 and 20 (inclusive).

df[df['score'].between(15, 20)]
```

```python
# 10. Write a Pandas program to select the rows where number of attempts in the examination 
# is less than 2 and score greater than 15.

ind = np.where((df['attempts'] <2 ) & (df['score'] >15))
df.iloc[ind]
```

```python
# 11. Write a Pandas program to change the score in row 'd' to 11.5.

df.loc['d','score'] = 11.5
df
```

```python
# 12. Write a Pandas program to calculate the sum of the examination attempts by the students.

from functools import reduce
add = reduce(lambda i,j: i+j, df['attempts'])
add
```

```python
# 13. Write a Pandas program to append a new row 'k' to data frame with given values for each column.

df.loc['k'] = ['Dibyendu', 17.2, 2, 'yes']
df
```

```python
# 14. Write a Pandas program to sort the DataFrame first by 'name' in descending order, 
# then by 'score' in ascending order.

df.sort_values(by=['name', 'score'], ascending=[True, False])
```

```python
# 15. Write a Pandas program to replace the 'qualify' column contains the values 
# 'yes' and 'no' with True and False.

df['qualify'] = df['qualify'].map({'yes':True, 'no':False})
df
```

```python
x = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(x, index=labels)
df
```

```python
# 16. Write a Pandas program to change the name 'James' to 'Suresh' in name column of the DataFrame.

ind = np.where(df['name'] == 'James')
index = df.iloc[ind]['name'].index

df.loc[index, 'name'] = ['Suresh']
df
```

```python
# 17. Write a Pandas program to iterate over rows in a DataFrame

exam_data = [{'name':'Anastasia', 'score':12.5}, 
             {'name':'Dima','score':9}, 
             {'name':'Katherine','score':16.5}]
ndf = pd.DataFrame(exam_data)

for index, row in ndf.iterrows():
    print(row['name'], row['score'])
```

```python
# 18. Write a Pandas program to rename columns of a given DataFrame

df.rename(columns={'name':'fname', 'score':'fscore', 'attempts':'fattempts'}, inplace=True)
df
```

```python
# 19. Write a Pandas program to select rows from a given DataFrame based on values in some columns.

nm = 'Emily'
ind = np.where(df['fname'] == nm)
df.iloc[ind].loc['e']
```

```python
# 20. Write a Pandas program to change the order of a DataFrame columns.

dct = {'col1':[10, 20, 30], 'col2':[11, 12, 13], 'col3':[21, 22, 23]}
df = pd.DataFrame(dct)
print(df, '\n\n')
print(df[['col3', 'col2', 'col1']])
```
