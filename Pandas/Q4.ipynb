# Pandas String and Regular Expression

```python
import numpy as np
import pandas as pd
```

```python
# 1. Write a Pandas program to convert all the string values to upper, lower cases 
# in a given pandas series. Also find the length of the string values.

ser = pd.Series(['arko', 'Dibyendu', 'raM', 'ShyaM'])
up = pd.Series([i.upper() for i in ser])
lw = pd.Series([i.lower() for i in ser])
ln = pd.Series(ser.str.len())

print(ser, '\n')
print(up, '\n')
print(lw, '\n')
print(ln)
```

```python
# 2. Write a Pandas program to remove whitespaces, left sided whitespaces and right sided 
# whitespaces of the string values of a given pandas series.

lst = [' arko  ', '   dibyendu ',  '  ram  ', '  ab c    ']
ser = pd.Series(lst)
print(ser.str.strip(), '\n')
print(ser.str.lstrip(), '\n')
print(ser.str.strip(), '\n')
```

```python
# 3. Write a Pandas program to add leading zeros to the integer column in a pandas series 
# and makes the length of the field to 8 digit.

nums = {'amount': [10, 250, 3000, 40000, 500000]}
print("Original dataframe:")
df = pd.DataFrame(nums)
print(df)
print("\nAdd leading zeros:")
df['amount'] = df['amount'].apply(lambda x: '{0:0>8}'.format(x))
print(df)
```

```python
# 4.  Write a Pandas program to capitalize all the string values of specified columns of 
# a given DataFrame. 

df = pd.DataFrame({
    'name': ['alberto','gino','ryan', 'Eesha', 'syed'],
    'date_of_birth ': ['17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [18.5, 21.2, 22.5, 22, 23]
})
# df['name'] = [n.capitalize() for n in df['name']]
df['name'] = list(map(lambda x: x.capitalize(), df['name']))
df
```

```python
# 5. Write a Pandas program to count of occurrence of a specified substring in a DataFrame column.


df = pd.DataFrame({
    'name_code': ['c001','c002','c022', 'c2002', 'c2222'],
    'date_of_birth ': ['12/05/2002','16/02/1999','25/09/1998','12/02/2022','15/09/1997'],
    'age': [18.5, 21.2, 22.5, 22, 23]
})
df['count'] = list(map(lambda x:x.count('2'), df['name_code']))
df
```

```python
# 6. Write a Pandas program to replace arbitrary values with other values in a given DataFrame.


df = pd.DataFrame({
    'company_code': ['A','B', 'C', 'D', 'A'],
    'date_of_sale': ['12/05/2002','16/02/1999','25/09/1998','12/02/2022','15/09/1997'],
    'sale_amount': [12348.5, 233331.2, 22.5, 2566552.0, 23.0]
})
df = df.replace("A", "C")
df
```

```python
# 7. Write a Pandas program to replace more than one value with other values in a given DataFrame.

df = df.replace(["B", "D"], ["X", "Y"])
df
```

```python
# 8. Write a Pandas program to replace more than one value with other values in a given DataFrame.


df = pd.DataFrame({
    'name': ['Alberto  Franco','Gino Ann Mcneill','Ryan  Parkes', 'Eesha Artur Hinton', 'Syed  Wharton'],
    'date_of_birth ': ['17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
    'age': [18.5, 21.2, 22.5, 22, 23]
})
df[["first", "middle", "last"]] = df["name"].str.split(" ", expand = True)
print("\nNew DataFrame:")
df
```

```python
# 9. Write a Pandas program to extract year between 1800 to 2200 from the specified 
# column of a given DataFrame. 

df = pd.DataFrame({
    'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
    'year': ['year 1800','year 1700','year 2300', 'year 1900', 'year 2200']})
df[['year', 'year']] = df['year'].str.split(" ", expand=True)
df['year'] = df['year'].astype(int)

ind = np.where((df['year'] >= 1800) & (df['year'] <= 2200))
df.iloc[ind]
```

```python
# 10. Write a Pandas program to extract date (format: mm-dd-yyyy) from a given column 
# of a given DataFrame.

df = pd.DataFrame({
    'company_code': ['Abcd','EFGF', 'zefsalf', 'sdfslew', 'zekfsdf'],
    'date_of_sale': ['12/05/2002','16/02/1999','05/09/1998','12/02/2022','15/09/1997'],
    'sale_amount': [12348.5, 233331.2, 22.5, 2566552.0, 23.0]
})
ind = np.where(df['date_of_sale'] == '16/02/1999')
df.iloc[ind]
```
